```python
import pandas as pd
import numpy as np
from talib.abstract import *
```

```python
pd.read_excel("BTC-USD.xlsx").head()
BTC = pd.read_excel("BTC-USD.xlsx")
BTC_modify=BTC.dropna()
BTC_modify=BTC_modify.reset_index(drop = True)
BTC_modify
```

```python
inputs = {
    'open':BTC_modify["Open"].to_numpy(),
    'high':BTC_modify["High"].to_numpy(),
    'low': BTC_modify["Low"].to_numpy(),
    'close':BTC_modify["Close"].to_numpy(),
    'volume':BTC_modify["Volume"].to_numpy(),
    'penetration':0
}
```

```python
def calculate_CDL(x):
    y=eval(x)
    BTC_modify[x]=0.0
    BTC_modify[x]=y(inputs)
```

```python
columns_CDL=['CDL2CROWS','CDL3BLACKCROWS','CDL3INSIDE',
            'CDL3LINESTRIKE','CDL3OUTSIDE','CDL3STARSINSOUTH',
            'CDL3WHITESOLDIERS','CDLABANDONEDBABY','CDLADVANCEBLOCK',
            'CDLBELTHOLD','CDLBREAKAWAY','CDLCLOSINGMARUBOZU',
            'CDLCONCEALBABYSWALL','CDLCOUNTERATTACK','CDLDARKCLOUDCOVER',
            'CDLDOJI','CDLDOJISTAR','CDLDRAGONFLYDOJI',
            'CDLENGULFING','CDLEVENINGDOJISTAR','CDLEVENINGSTAR',
            'CDLGAPSIDESIDEWHITE','CDLGRAVESTONEDOJI','CDLHAMMER',
            'CDLHANGINGMAN','CDLHARAMI','CDLHARAMICROSS',
            'CDLHIGHWAVE','CDLHIKKAKE','CDLHIKKAKEMOD',
            'CDLHOMINGPIGEON','CDLIDENTICAL3CROWS','CDLINNECK',
            'CDLINVERTEDHAMMER','CDLKICKING','CDLKICKINGBYLENGTH','CDLLADDERBOTTOM',
            'CDLLONGLEGGEDDOJI','CDLLONGLINE','CDLMARUBOZU',
            'CDLMATCHINGLOW','CDLMATHOLD','CDLMORNINGDOJISTAR',
            'CDLMORNINGSTAR','CDLONNECK','CDLPIERCING',
            'CDLRICKSHAWMAN','CDLRISEFALL3METHODS','CDLSEPARATINGLINES',
            'CDLSHOOTINGSTAR','CDLSHORTLINE','CDLSPINNINGTOP',
            'CDLSTALLEDPATTERN','CDLSTICKSANDWICH','CDLTAKURI',
            'CDLTASUKIGAP','CDLTHRUSTING','CDLTRISTAR',
            'CDLUNIQUE3RIVER','CDLUPSIDEGAP2CROWS','CDLXSIDEGAP3METHODS']
```

```python
BTC_trade=BTC_final.iloc[:,0:6]
BTC_trade
```

```python
BTC_trade['Average']=0.0
for i in range(len(BTC_final)):
    BTC_trade['Average'][i]=BTC_final.iloc[i,7:len(BTC_final)].mean()
BTC_trade
```

```python
BTC_trade['Buy']=0
BTC_trade['Hold']=0
BTC_trade['Sell']=0
BTC_trade
```

```python
for i in range(1, len(BTC_trade)):
    if BTC_trade.loc[i,'Average'] <0 and BTC_trade.loc[i-1,'Hold']==0:
        BTC_trade['Buy'][i]=1
        BTC_trade['Hold'][i]=1
        BTC_trade['Sell'][i]=0
    elif BTC_trade.loc[i,'Average'] > 0 and BTC_trade.loc[i-1,'Hold']==1:
        BTC_trade['Buy'][i]=0
        BTC_trade['Hold'][i]=0
        BTC_trade['Sell'][i]=1
    elif(BTC_trade.loc[i-1,'Hold']>0):
        BTC_trade['Buy'][i]=0
        BTC_trade['Hold'][i]=1
        BTC_trade['Sell'][i]=0
BTC_trade
```

```python
BTC_trade["Number"]=0.0
BTC_trade["Value"]=0.0
BTC_trade["Return"]=0.0
BTC_trade.head(20)
```

```python
initial = 10000
for i in range(5, len(BTC_trade)):
    if(BTC_trade['Buy'][i] == 1)and(BTC_trade['Hold'][i]== 1)and(BTC_trade['Sell'][i] == 0):#buy
        #持有数量=手中资本/标的的收盘价
        BTC_trade['Number'][i] = initial/BTC_trade['Adj Close'][i]
        #市值 = 持有數量* 該標的收盤價
        BTC_trade['Value'][i] = BTC_trade['Number'][i] * BTC_trade['Adj Close'][i]
        #期間報酬 = (市值-投入成本)/投入成本
        BTC_trade['Return'][i] = round((BTC_trade['Value'][i]-initial)/initial,4)
    elif(BTC_trade['Buy'][i] == 0)and(BTC_trade['Hold'][i] == 0)and(BTC_trade['Sell'][i] == 1):#sell
        #持有數量為 0 
        BTC_trade['Number'][i] = 0
        #市值 = 持有數量* 該標的收盤價
        BTC_trade['Value'][i] = BTC_trade['Number'][i-1] * BTC_trade['Adj Close'][i]
        #期間報酬 = (市值-投入成本)/投入成本
        BTC_trade['Return'][i] = round((BTC_trade['Value'][i]-initial)/initial,4)
    elif(BTC_trade['Buy'][i] == 0)and(BTC_trade['Hold'][i]==1)and(BTC_trade['Sell'][i] == 0):#hold
        BTC_trade['Number'][i] = BTC_trade['Number'][i-1]
        BTC_trade['Value'][i] = BTC_trade['Number'][i] * BTC_trade['Adj Close'][i]
        BTC_trade['Return'][i] = round((BTC_trade['Value'][i]-initial)/initial,4)
BTC_trade
```

```python
BTC_trade["Holding_days"]=0.0
BTC_trade["Annualized_Return"]=0.0
BTC_trade
```

```python
start = 0
end = 0
for i in range(1, len(BTC_trade)):
    if(BTC_trade['Buy'][i] == 1):
        BTC_trade['Holding_days'][i] = 1
        start = BTC_trade['Date'][i]
    elif((BTC_trade['Hold'][i] == 1) or (BTC_trade['Sell'][i] == 1))and(BTC_trade['Buy'][i] == 0):
        end = BTC_trade['Date'][i]
        delta = (end - start).days
        BTC_trade['Holding_days'][i]= delta+1
BTC_trade.head(20) 
```

```python
for i in range(1, len(BTC_trade)):
    if(BTC_trade['Buy'][i]==1)or(BTC_trade['Hold'][i]==1)or(BTC_trade['Sell'][i]==1):
        BTC_trade['Annualized_Return'][i] = (1+BTC_trade['Return'][i])**(365/BTC_trade['Holding_days'][i])-1
BTC_trade
```

```python
# annual return computation
# https://www.investopedia.com/terms/a/annualized-total-return.asp
period_return = 1
annual_return = 1
trading_counts = 0
for i in range(1, len(BTC_trade)):
    if(BTC_trade['Sell'][i]==1):
        period_return = period_return * (1+BTC_trade['Return'][i])
        annual_return = annual_return * (1+BTC_trade['Annualized_Return'][i])
        trading_counts = trading_counts + 1
print('trading_counts: ',trading_counts)
final_period_return = period_return**(1/trading_counts)-1
print('final_period_return:',final_period_return)
final_annual_return = annual_return**(1/trading_counts)-1
print('final_annual_return:',final_annual_return)
```
